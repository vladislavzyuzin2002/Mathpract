1) Plot the populations for the years in the period. Add legend and axis labels to the plot.

2) Find which years any of the populations is above 50000. Show time periods where populations are above 50000 in the plot of populations.

3) Find which year each species had the largest population. Mark these years in the plot of populations.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

fig = plt.figure(figsize=(20, 10), dpi=80)
fig.patch.set_facecolor('xkcd:white')

S = np.loadtxt('http://www.scipy-lectures.org/_downloads/populations.txt')
X = S[:,0]
P = np.array([50000]*21)
plt.plot(X, S[:,1], label='Динамика изменения численности зайцев', c='r', linestyle='-', linewidth=2)
plt.plot(X, S[:,2], label='Динамика изменения численности рысей', c='b', linestyle='-', linewidth=2)
plt.plot(X, S[:,3], label='Динамика изменения количества моркови', c='g', linestyle='-', linewidth=2)
plt.scatter(X, S[:,1], c='r', edgecolor='k')
plt.scatter(X, S[:,2], c='b', edgecolor='k')
plt.scatter(X, S[:,3], c='g', edgecolor='k')
plt.scatter(X, S[:,1], c='r', edgecolor='k')

plt.plot(X, P, c='black', linestyle='--', linewidth=3)

plt.scatter(S[np.where(S[:,1] == np.max(S[:,1]))][0][0], np.max(S[:,1]), c='red', marker="*", linewidths=10)
plt.scatter(S[np.where(S[:,2] == np.max(S[:,2]))][0][0], np.max(S[:,2]), c='b', marker="*", linewidths=10)
plt.scatter(S[np.where(S[:,3] == np.max(S[:,3]))][0][0], np.max(S[:,3]), c='g', marker="*", linewidths=10)
plt.xlabel('Год', fontsize=20)
plt.ylabel('Количество')
plt.xticks(fontsize=20)
plt.legend()
plt.show()
```

4) Assuming total population of hares and lynxes is 100%, plot stacked bar graph showing ratio of each specis for the years in the period.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(40, 10), dpi=100)
data_1 = S[:,1] / (S[:,1] + S[:,2]) *100
data_2 = S[:,2] / (S[:,1] + S[:,2])*100

ax.bar(X, data_1)
ax.bar(X, data_2, bottom = data_1)

fig.set_figwidth(15)
fig.set_figheight(6)
fig.patch.set_facecolor('xkcd:white')
plt.legend(['Зайцы','Рыси'])
plt.xlabel("Год")
plt.ylabel("Проценты")

plt.show()
```

5) Find the mean and std of the populations of each species. Plot the historgam of population for each species and show mean values with vertical line. Arrange 3 subplots in a row.

```python
fig = plt.figure(figsize=(20, 10), dpi=80)
fig.patch.set_facecolor('xkcd:white')
Z1 = plt.bar(X, S[:,1])
plt.xticks(X + 0.4, Z1)
plt.plot(X, np.array([float(sum(S[:,1])) / max(len(S[:,1]), 1)]*21), c='black', linestyle='--', linewidth=3)
plt.xlabel('Год')
plt.ylabel('Количество')
plt.title('Гистограмма для зайцев')
plt.show()

fig = plt.figure(figsize=(20, 10), dpi=80)
fig.patch.set_facecolor('xkcd:white')
Z2 = plt.bar(X, S[:,2], facecolor='g')
plt.xticks(X + 0.4, Z2)
plt.plot(X, np.array([float(sum(S[:,2])) / max(len(S[:,2]), 1)]*21), c='black', linestyle='--', linewidth=3)
plt.xlabel('Год')
plt.ylabel('Количество')
plt.title('Гистограмма для рысей')
plt.show()

fig = plt.figure(figsize=(20, 10), dpi=80)
fig.patch.set_facecolor('xkcd:white')
Z3 = plt.bar(X, S[:,3], facecolor='r')
plt.xticks(X + 0.4, Z3)
plt.plot(X, np.array([float(sum(S[:,3])) / max(len(S[:,3]), 1)]*21), c='black', linestyle='--', linewidth=3)
plt.xlabel('Год')
plt.ylabel('Количество')
plt.title('Гистограмма для моркови')
plt.show()
```
